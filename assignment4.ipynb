```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("https://github.com/AnjulaMehto/MCA/blob/45704cf74a200ca2100f013bedd4b4919d5e055d/company_sales_data.csv?raw=true")
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data["month_number"],data["total_profit"],label="total profit by month",marker="o")
plt.legend()
plt.xlabel("month")
plt.ylabel("profit")
```

```python
plt.figure(figsize=(12,6))
plt.plot(data["month_number"],data["facecream"],label="facecream")
plt.plot(data["month_number"],data["facewash"],label="facewash")
plt.plot(data["month_number"],data["toothpaste"],label="toothpaste")
plt.plot(data["month_number"],data["bathingsoap"],label="bathing soap")
plt.plot(data["month_number"],data["shampoo"],label="shampoo")
plt.plot(data["month_number"],data["moisturizer"],label="moisturizer")
plt.legend()
plt.xlabel("month")
plt.ylabel("profit")
```

```python
plt.figure(figsize=(12,6))

plt.subplot(2,1,1)
features=["january","feburary","march","april","may","june","july","august","september","october","november","december"]
values=data["total_profit"].values
plt.bar(features,values)
plt.xlabel("month")
plt.ylabel("total profit")

plt.subplot(2,1,2)
features=["facecream","facewash","toothpaste","bathingsoap","shampoo","moistureizer"]
values=[]
for i in range(1,7):
  values.append(data.iloc[:,i].sum())
plt.bar(features,values)
plt.xlabel("product")
plt.ylabel("overall count")
```

```python
plt.figure(figsize=(12,6))
plt.scatter(data["month_number"],data["facecream"],label="facecream",marker=".")
plt.scatter(data["month_number"],data["facewash"],label="facewash",marker="*")
plt.scatter(data["month_number"],data["toothpaste"],label="toothpaste",marker="+")
plt.scatter(data["month_number"],data["bathingsoap"],label="bathing soap",marker="o")
plt.legend()
plt.xlabel("month")
plt.ylabel("profit")
```
